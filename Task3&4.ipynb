```python
import numpy as np

x = np.linspace(-10, 10, 200)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, label="sigmoid(x)", color="red", linewidth=2)

plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")

plt.legend()

plt.savefig("Task4.png")
```
